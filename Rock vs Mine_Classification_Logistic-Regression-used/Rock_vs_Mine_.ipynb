```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#data reading
sonar_data=pd.read_csv('/content/Sonar Data.csv',header=None)
```

```python
#by default when head is called, first 5 rows are displayed
sonar_data.head()
```

```python
# gives number of rows and columns
sonar_data.shape
```

```python
#It gives statistical data
sonar_data.describe()
```

```python
#It gives count of the values in the particular column if mentioned
sonar_data[60].value_counts()
```

```python
# It groups by the values in the particular column and takes average of the column it is associated with
sonar_data.groupby(60).mean()
```

```python
#separating data and labels
X = sonar_data.drop(columns=60, axis=1) #if we are dropping a row, then azis should be equal to 0
Y=sonar_data[60]
```

```python
# "test_size" parameter gives the percentage of the test data to be alloted from the whole dataset available. Here 0.1 means 10% of the whole dataset given to us
# "Stratify" this parameter allows us to allot the final output in equal ratio whenever it is split into train and test data set.
#"random_state" is basically used to reproduce the same "data points" selected. It may be 1 or 2 or 3, but given the number it will behave the same way.
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.1, stratify =Y, random_state=1)
```

```python
print(X_train.shape,X_test.shape)
```

```python
model =LogisticRegression()
```

```python
# training the data using X_train and Y_train
model.fit(X_train,Y_train)
```

```python
# Getting the accuracy score for the train data
X_train_predictions=model.predict(X_train)
train_accuracy=accuracy_score(X_train_predictions,Y_train)
```

```python
print("Accuracy of training data is :", train_accuracy)
```

```python
# Getting the accuracy score for the train data
X_test_predictions=model.predict(X_test)
test_accuracy=accuracy_score(X_test_predictions,Y_test)
```

```python
print("Accuracy of test data is :", test_accuracy)
```
